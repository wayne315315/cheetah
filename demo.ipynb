```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from database import load_tw_stock


df = load_tw_stock() # this may take a while; might run out of memory
```

```python
df.head()
```

```python
df.tail()
```

```python
code1 = "0050"
code2 = "2330"
col = "Closing Price"

df[col].loc[code1].plot(color="red")
df[col].loc[code2].plot(color="blue")
red_patch = mpatches.Patch(color='red', label=code1)
blue_patch = mpatches.Patch(color='blue', label=code2)
plt.legend(handles=[red_patch, blue_patch])
plt.title(col)
plt.show()
```

```python
start = "2004-01-01"
end = "2010-01-01"
df[col].loc[code1].loc[start:end].plot(color="red")
df[col].loc[code2].loc[start:end].plot(color="blue")
red_patch = mpatches.Patch(color='red', label=code1)
blue_patch = mpatches.Patch(color='blue', label=code2)
plt.legend(handles=[red_patch, blue_patch])
plt.title(col)
plt.show()
```
