```python
import sqlite3

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from config import DB

conn = sqlite3.connect(DB)
df = pd.read_sql_query("SELECT * FROM tw_stock", conn, index_col=["Security Code", "Date"] , parse_dates=["Date"])
```

```python
code1 = "0050"
code2 = "2330"
col = "Closing Price"

start = pd.Timestamp(year=2006, month=3, day=1)
end = pd.Timestamp(year=2006, month=4, day=1)
df[col].loc[code1].plot(color="red")
df[col].loc[code2].plot(color="blue")
red_patch = mpatches.Patch(color='red', label=code1)
blue_patch = mpatches.Patch(color='blue', label=code2)
plt.legend(handles=[red_patch, blue_patch])
plt.title(col)
plt.show()
```
